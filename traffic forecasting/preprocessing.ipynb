```python
for col in train.columns[2:]:
    x1 = train.loc[(train.날짜 == 20200206) & (train.시간 == 17), col]
    x2 = train.loc[(train.날짜 == 20200206) & (train.시간 == 19), col]
    train.loc[(train.날짜 == 20200206) & (train.시간 == 18), col] = (x1.values + x2.values) / 2
```

```python
x1 = train.loc[(train.날짜 == 20200222) & (train.시간 > 12)].iloc[:,2:].reset_index(drop=True)
x2 = train.loc[(train.날짜 == 20200307) & (train.시간 > 12)].iloc[:,2:].reset_index(drop=True)
x = (x1 + x2) / 2
```

```python
data = 20200229
for time in range(13, 24):
    new_data = {}
    new_data['날짜'] = data
    new_data['시간'] = time
    for col in train.columns[2:]:
        new_data[col] = x[col].iloc[time - 13]
    
    train = train.append(new_data,ignore_index=True)
```
