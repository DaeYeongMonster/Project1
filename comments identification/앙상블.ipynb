```python
import pandas as pd
import os
import numpy as np
```

```python
bert_large = pd.read_csv(os.path.join('/content/drive/MyDrive/Colab Notebooks/traffic forecasting/Project1/comments identification/result/','submission_kcbert_large_proba.csv'))
f1 = pd.read_csv(os.path.join('/content/drive/MyDrive/Colab Notebooks/traffic forecasting/Project1/comments identification/result/','submission_change f1 score_proba.csv'))
f1_google_pororo = pd.read_csv(os.path.join('/content/drive/MyDrive/Colab Notebooks/traffic forecasting/Project1/comments identification/result/','submission_f1 score, google, pororo_proba.csv'))
```

```python
x1 = bert_large['proba']
x2 = f1['proba']
x3 = f1_google_pororo['proba']
```

```python
label = []
for j in range(len(x1)):
    x11 = []
    for i in x1[j].split(',',-1):
        if (i[0] == '[') or (i[0] == ' '):
            i = i[1:]
        if i[-1] == ']':
            i = i[:-1]
        i = float(i)
        x11.append(i)
    x22 = []
    for i in x2[j].split(',',-1):
        if (i[0] == '[') or (i[0] == ' '):
            i = i[1:]
        if i[-1] == ']':
            i = i[:-1]
        i = float(i)
        x22.append(i)
    x33 = []
    for i in x3[j].split(',',-1):
        if (i[0] == '[') or (i[0] == ' '):
            i = i[1:]
        if i[-1] == ']':
            i = i[:-1]
        i = float(i)
        x33.append(i)

    x11 = np.array(x11)
    x22 = np.array(x22)
    x33 = np.array(x33)

    label.append((x11 + x22 + x33).argmax())

label
```

```python
# 0-5 사이의 라벨 값 별로 bias, hate로 디코딩 하기 위한 딕셔너리
bias_dict = {0: 'none', 1: 'none', 2: 'others', 3:'others', 4:'gender', 5:'gender'}
hate_dict = {0: 'none', 1: 'hate', 2: 'none', 3:'hate', 4:'none', 5:'hate'}

# 인코딩 값으로 나온 타겟 변수를 디코딩
pred_bias = ['' for i in range(len(label))]
pred_hate = ['' for i in range(len(label))]

for idx, x in enumerate(label):
    pred_bias[idx]=(str(bias_dict[x]))
    pred_hate[idx]=(str(hate_dict[x]))
print('decode Completed!')
```
