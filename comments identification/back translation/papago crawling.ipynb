```python
from selenium import webdriver
import time

driver1 = webdriver.Chrome('./chromedriver/chromedriver.exe')
driver2 = webdriver.Chrome('./chromedriver/chromedriver.exe')

# 파파고 접속
driver1.get('https://papago.naver.com/?sk=ko&tk=en')
driver2.get('https://papago.naver.com/?sk=en&tk=ko&hn=0')
```

```python
import pandas as pd
```

```python
train_df = pd.read_csv('/자료/뜸부기/2021년/이어드림 프로젝트/모의 경진 대회/comments identification/data/train.csv')
```

```python
train_df.title
```

```python
chinese = {
'故': '고', 
'無': '무',
'美': '미',
'韓': '한',
'中': '중',
'日': '일',
'父': '부',
'新': '신',
'母': '모',
'多': '다',
'行': '행',
'前': '전',
'軍': '군',
'家': '가',
'史': '사',
'男': '남',
'子': '자',
'大': '대',
'心': '심', 
'人': '인',
'女': '여', 
'眞': '진',
'愛': '애',
'色': '색',
'性': '성',
'役': '역',
'神': '신',
'有': '유',
'勝': '승',
'伊': '이',
'檢': '검',
'女': '녀'
}
```

```python
for char in chinese:
    train_df["title"] = train_df["title"].str.replace(pat=char, repl=chinese[char], regex=True)
```

```python
train_df["title"] = train_df["title"].str.replace(pat=r'\[(.{1,10})\]', repl=r'', regex=True)
train_df["title"] = train_df["title"].str.replace(pat=r'\((.{1,10})\)', repl=r'', regex=True)
train_df["title"] = train_df["title"].str.replace(pat=r'ing', repl=r'중', regex=True)
train_df["title"] = train_df["title"].str.replace(pat=r'섹션TV', repl=r'', regex=True)
```

```python
train_df["title"] = train_df["title"].str.replace(pat=r'[^\w]', repl=r' ', regex=True)
```

```python
train_df["title"] = train_df["title"].str.replace(pat=r' {2,10}', repl=r' ', regex=True)
train_df["title"] = train_df["title"].str.replace(pat=r'^ ', repl=r'', regex=True)
train_df["title"] = train_df["title"].str.replace(pat=r' $', repl=r'', regex=True)
```

```python
train_df["title"] = train_df["title"].str.replace(pat=r'다$', repl=r'다.', regex=True)
```

```python
back_tran_title = []
for index, tran in enumerate(papago['papago title']):
    if index % 100 == 0:
        print(index)
        
    if len(tran) < 10:
        string = papago['title'].iloc[index]
        try:
            driver1.find_element_by_xpath('//*[@id="sourceEditArea"]/button').click()
        except:
            pass
        driver1.find_element_by_xpath('//*[@id="sourceEditArea"]').send_keys('1' + string) # 첫 입력 글짜가 짤리는 현상있었서 1을 추가했다.
        driver1.find_element_by_css_selector('#btnTranslate').click()
        time.sleep(10)
        transResult = driver1.find_element_by_xpath('//*[@id="txtTarget"]').text

        try:
            driver2.find_element_by_xpath('//*[@id="sourceEditArea"]/button').click()
        except:
            pass
        driver2.find_element_by_xpath('//*[@id="sourceEditArea"]').send_keys('1' + transResult)
        driver2.find_element_by_css_selector('#btnTranslate').click()
        time.sleep(10)
        back_tran_title.append(driver2.find_element_by_xpath('//*[@id="txtTarget"]').text)
    else:
        back_tran_title.append(tran)
```
