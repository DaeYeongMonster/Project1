```python
# test 전처리
test_df =  pd.read_csv('/content/drive/MyDrive/Colab Notebooks/traffic forecasting/Project1/comments identification/data/test.csv')
```

```python
chinese = {
'故': '고', 
'無': '무',
'美': '미',
'韓': '한',
'中': '중',
'日': '일',
'父': '부',
'新': '신',
'母': '모',
'多': '다',
'行': '행',
'前': '전',
'軍': '군',
'家': '가',
'史': '사',
'男': '남',
'子': '자',
'大': '대',
'心': '심', 
'人': '인',
'女': '여', 
'眞': '진',
'愛': '애',
'色': '색',
'性': '성',
'役': '역',
'神': '신',
'有': '유',
'勝': '승',
'伊': '이',
'檢': '검',
'女': '녀'
}
```

```python
for char in chinese:
    test_df["title"] = test_df["title"].str.replace(pat=char, repl=chinese[char], regex=True)
```

```python
test_df["title"] = test_df["title"].str.replace(pat=r'\[(.{1,10})\]', repl=r'', regex=True)
test_df["title"] = test_df["title"].str.replace(pat=r'\((.{1,10})\)', repl=r'', regex=True)
test_df["title"] = test_df["title"].str.replace(pat=r'ing', repl=r'중', regex=True)
test_df["title"] = test_df["title"].str.replace(pat=r'섹션TV', repl=r'', regex=True)
```

```python
test_df["title"] = test_df["title"].str.replace(pat=r'[^\w]', repl=r' ', regex=True)
```

```python
test_df["title"] = test_df["title"].str.replace(pat=r' {2,10}', repl=r' ', regex=True)
test_df["title"] = test_df["title"].str.replace(pat=r'^ ', repl=r'', regex=True)
test_df["title"] = test_df["title"].str.replace(pat=r' $', repl=r'', regex=True)
```

```python
test_df["title"] = test_df["title"].str.replace(pat=r'다$', repl=r'다.', regex=True)
```

```python
test_df.to_csv('/content/drive/MyDrive/Colab Notebooks/traffic forecasting/Project1/comments identification/data/back taranslation test.csv', index=False)
```
